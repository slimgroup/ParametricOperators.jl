```julia
using Base64
using LinearAlgebra
using Random
using Revise
using ParametricOperators
using Printf
using PyCall
using Zygote

@pyimport matplotlib.animation as anim
using PyPlot
```

```julia
# Global parameters
N = 2     # Number of exponentials
δ = 0.1   # Spacing between observations
t = 0:δ:1 # Observation points
t = collect(t)

# Model given by sum of exponentials
f(c, α; x=t) = sum(map(i -> c[i].*exp.(α[i].*x), 1:N))

# Create ground truth f(x) and noisy observations y
Random.seed!(123)
αt = 10*randn(N)
ct = rand(N)
σ  = 0.05
y  = f(ct, αt) + σ.*randn(length(t))

@show αt ct

figure()
scatter(t, y, marker="x", color="red")
```

```julia
# Setup parametric function (linear in first argument, nonlinear in second argument)
Φ = ParLinearFunction(f, N; P=Parametric)
```

```julia
# Inner solve for linear parameters using Nesterov's method
function ĉ(A, c=rand(N); η=0.01, max_iter=10000, tol=1e-5)
    c_prev = zero(c)
    for i = 1:max_iter
        
        # Nesterov momentum
        p = (i-1)/(i+2).*(c .- c_prev)
        
        # Compute the gradient w.r.t c plus momentum
        # ∂/∂c ||y - A*(c+p)||₂
        g = gradient(x -> norm(y - A*(x.+p)), c)[1]
        if (norm(g)) ≤ tol
            break
        end
        
        # Nesterov update cₖ₊₁ = cₖ + pₖ - η*∇f(cₖ+pₖ)
        c .+= p
        c .-= η.*g
        c_prev .= c
    end
    return c
end
```

```julia
# Outer solve for nonlinear parameters
α = randn(N)
η = 0.01
max_iter = 2000
tol = 1e-3

# Store iterates
cs = []
αs = []
rs = []

for i = 1:max_iter
    
    # Solve the convex problem for c at the parameters given by α
    # min ||Φ(α)*c - y||₂ w.r.t c
    c = ĉ(Φ(α))
        
    # Compute the jacobian w.r.t α on the residual r = y - Φ(α)*c
    J = jacobian(a -> y - Φ(a)*c, α)[1]
    r = y - Φ(α)*c
    
    # Step using Jᵀr
    g = J'*r
    α .-= g
    
    push!(cs, c)
    push!(αs, copy(α))
    push!(rs, r)
    
    # Display iterates
    if mod1(i, 100) == 1
        iterate = lpad(i,4,"0")
        println("i = $iterate, norm(y - Φ(a)*c(α)) = $(norm(r))")
    end
end
```

```julia
t̂ = 0:0.01:1
chat = cs[end]
αhat = αs[end]
ŷ = f(chat, αhat; x=t̂)

figure(figsize=(6, 6))
scatter(t, y, marker="x", color="red", label=L"\mathbf{y}_{obs}")
plot(t̂, ŷ, label=L"\hat{\mathbf{y}}")
xlabel(L"t")
ylabel(L"y")
title(latexstring("Solution ", L"\mathbf{\Phi}(\mathbf{\alpha})\mathbf{c} = \sum_{i=1}^N c_i e^{\alpha_i t}"))
legend()

figure(figsize=(14, 4))

step = 100
res = collect(map(r -> norm(r), rs))
subplot(131)
plot(collect(1:length(res)), res)
scatter(1:step:length(res), res[1:step:end], marker="o", s = 10)
yscale("log")
xlabel(latexstring("Iteration ", L"k"))
ylabel(L"||\mathbf{y} - \Phi(\mathbf{\alpha})\mathbf{c}||_2^2")
title("Convergence of residual")

cs_out = hcat(cs...)
subplot(132)
plot(cs_out[1,:], cs_out[2,:], label=L"\mathbf{c}_k")
scatter(cs_out[1,1:step:end-1], cs_out[2,1:step:end-1], marker="o", s = 10)
scatter(cs_out[1,end], cs_out[2,end], marker="x", s = 20, color="red", label=L"\mathbf{\hat{c}}")
xlabel(L"c_1")
ylabel(L"c_2")
title(latexstring("Convergence of ", L"\mathbf{c}"))
legend()

αs_out = hcat(αs...)
subplot(133)
plot(αs_out[1,:], αs_out[2,:], label=L"\mathbf{\alpha}_k")
scatter(αs_out[1,1:step:end-1], αs_out[2,1:step:end-1], marker="o", s = 10)
scatter(αs_out[1,end], αs_out[2,end], marker="x", s = 20, color="red", label=L"\mathbf{\hat{\alpha}}")
xlabel(L"\alpha_1")
ylabel(L"\alpha_2")
title(latexstring("Convergence of ", L"\mathbf{\alpha}"))
legend()

tight_layout()
```

```julia
# https://genkuroki.github.io/documents/Jupyter/20170624%20Examples%20of%20animations%20in%20Julia%20by%20PyPlot%20and%20matplotlib.animation.html#Example-3
function showanim(filename)
    base64_video = base64encode(open(filename))
    display("text/html", """<video controls src="data:video/x-m4v;base64,$base64_video">""")
end

fig = figure(figsize=(6,6))

# i=0,1,...,frames-1
function animate(i)
    clf()
    t̂ = 0:0.01:1
    chat = cs[i+1]
    αhat = αs[i+1]
    ŷ = f(chat, αhat; x=t̂)
    scatter(t, y, marker="x", color="red", label=L"\mathbf{y}_{obs}")
    xlim((-0.1, 1.1))
    ylim((-0.25, 1.25))
    xlabel(L"t")
    ylabel(L"y")
    title(latexstring("Solution ", L"\mathbf{\Phi}(\mathbf{\alpha})\mathbf{c} = \sum_{i=1}^N c_i e^{\alpha_i t}", " at iteration $(i+1)"))
    plot(t̂, ŷ, label=L"\hat{\mathbf{y}}")
end

nframes = length(rs)
frametol = 1e-4
for i = 1:length(rs)-1
    if norm(rs[i+1].-rs[i]) ≤ frametol
        nframes = i
        break
    end
end

# Warning: takes a long time!
withfig(fig) do
    myanim = anim.FuncAnimation(fig, animate, frames=nframes, interval=25, blit=true)
    myanim[:save]("convergence.mp4", bitrate=-1, extra_args=["-vcodec", "libx264", "-pix_fmt", "yuv420p"])
end

showanim("convergence.mp4")
```

```julia
using ImageIO
using LaTeXStrings
using Revise
using ParametricOperators
```

```julia
# Define two three-dimensional separable operators
Ax = ParMatrix(ComplexF64, 8, 16);
Ay = ParMatrix(ComplexF64, 8, 14);
Az = ParMatrix(ComplexF64, 8, 12);

Fx = ParDFT(ComplexF64, 16);
Fy = ParDFT(ComplexF64, 14);
Fz = ParDFT(ComplexF64, 12);

A = Az ⊗ Ay ⊗ Ax;
F = Fz ⊗ Fy ⊗ Fx;
C = A*F;
```

```julia
@show Domain(C)
@show Range(C)
@show complexity(C)
```

```julia
fname = plot_operator_graph(C; include_on_label=Dict("Domain" => Domain, "Range" => Range, "order" => ParametricOperators.order));
f = open(fname, "r")
s = read(f,String)
close(f)
display("image/svg+xml",s)
```

```julia
latexstring(latex_string(C))
```

```julia
ts = collect(transforms(C));
```

```julia
complexities = complexity.(ts);
min_complexity_idx = argmin(complexities);
max_complexity_idx = argmax(complexities);
```

```julia
C_min_complexity = ts[min_complexity_idx];
fname = plot_operator_graph(C_min_complexity; include_on_label=Dict("Domain" => Domain, "Range" => Range, "order" => ParametricOperators.order));
f = open(fname, "r")
s = read(f,String)
close(f)
display("image/svg+xml",s)
```

```julia
@show complexity(C_min_complexity)
latexstring(latex_string(C_min_complexity))
```

```julia
C_max_complexity = ts[max_complexity_idx];
fname = plot_operator_graph(C_max_complexity; include_on_label=Dict("Domain" => Domain, "Range" => Range, "order" => ParametricOperators.order));
f = open(fname, "r")
s = read(f,String)
close(f)
display("image/svg+xml",s)
```

```julia
@show complexity(C_max_complexity)
latexstring(latex_string(C_max_complexity))
```
